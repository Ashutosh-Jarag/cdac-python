# Series — 1D labeled array:

```python
import pandas as pd

# create series
s = pd.Series([10,20,30,40,50])
print(s)
# 0    10
# 1    20
# 2    30
# 3    40
# 4    50

# custom index
s = pd.Series([10,20,30], index=["a","b","c"])
print(s)
# a    10
# b    20
# c    30

# from dict
s = pd.Series({"name":"Rahul", "age":20, "city":"Pune"})
print(s)
# name    Rahul
# age        20
# city     Pune

# access
print(s["name"])   # Rahul
print(s[0])        # 10
print(s[1:3])      # slicing!
```

# 3. EXPLORING DATA:

```python
df = pd.DataFrame({
    "name"  : ["Rahul","Priya","John","Neha","Sam"],
    "age"   : [20, 21, 19, 22, 20],
    "marks" : [85, 92, 78, 88, 65],
    "city"  : ["Pune","Mumbai","Delhi","Pune","Mumbai"]
})

# basic info
print(df.head())        # first 5 rows
print(df.tail())        # last 5 rows
print(df.head(3))       # first 3 rows
print(df.shape)         # (5, 4) rows x cols
print(df.info())        # column types
print(df.describe())    # statistics
print(df.columns)       # column names
print(df.dtypes)        # data types
print(df.isnull())      # check null values
print(df.isnull().sum())# count nulls
print(len(df))          # row count
```

# 4. SELECTING DATA:

```python
# select column
print(df["name"])           # single column
print(df[["name","marks"]]) # multiple columns

# select row by index — iloc
print(df.iloc[0])      # first row
print(df.iloc[1:3])    # rows 1 to 2
print(df.iloc[0,1])    # row 0 col 1

# select row by label — loc
print(df.loc[0])           # row 0
print(df.loc[0,"name"])    # specific cell
print(df.loc[0:2,"name":"marks"])  # range

# conditional selection
print(df[df["marks"] > 80])           # marks > 80
print(df[df["city"] == "Pune"])       # city is Pune
print(df[(df["marks"]>80) & (df["city"]=="Pune")])  # multiple!
```

# 5. MODIFYING DATA:

```python
# add column
df["grade"] = ["A","A","B","A","C"]

# add calculated column
df["passed"] = df["marks"] >= 40

# update column
df["marks"] = df["marks"] + 5

# rename columns
df.rename(columns={"name":"Name", "marks":"Marks"}, inplace=True)

# drop column
df.drop(columns=["grade"], inplace=True)

# drop row
df.drop(index=0, inplace=True)

# update specific value
df.loc[0,"marks"] = 95
df.at[0,"marks"]  = 95
```

# 6. HANDLING NULL VALUES:

```python
import numpy as np

df = pd.DataFrame({
    "name"  : ["Rahul","Priya",None,"Neha"],
    "marks" : [85, None, 78, 88],
    "city"  : ["Pune","Mumbai",None,"Pune"]
})

# check nulls
print(df.isnull())
print(df.isnull().sum())

# drop nulls
df.dropna()              # drop rows with any null
df.dropna(axis=1)        # drop cols with any null

# fill nulls
df.fillna(0)             # fill with 0
df.fillna("unknown")     # fill with string
df["marks"].fillna(df["marks"].mean())  # fill with mean!
```

# 7. GROUPBY — group and aggregate:

```python
df = pd.DataFrame({
    "name"  : ["Rahul","Priya","John","Neha","Sam"],
    "city"  : ["Pune","Mumbai","Pune","Mumbai","Pune"],
    "marks" : [85, 92, 78, 88, 65]
})

# group by city
grouped = df.groupby("city")

print(grouped["marks"].mean())   # avg marks per city
print(grouped["marks"].sum())    # total marks per city
print(grouped["marks"].max())    # max marks per city
print(grouped["marks"].count())  # count per city

# multiple aggregations
print(grouped["marks"].agg(["mean","max","min","count"]))
```

# SORTING

# 9. MERGING & JOINING:

```python
df1 = pd.DataFrame({
    "id"  : [1,2,3],
    "name": ["Rahul","Priya","John"]
})

df2 = pd.DataFrame({
    "id"   : [1,2,4],
    "marks": [85,92,78]
})

# merge — like SQL JOIN
pd.merge(df1, df2, on="id")              # inner join
pd.merge(df1, df2, on="id", how="left")  # left join
pd.merge(df1, df2, on="id", how="right") # right join
pd.merge(df1, df2, on="id", how="outer") # outer join

# concat — stack dataframes
pd.concat([df1, df2])          # vertical
pd.concat([df1, df2], axis=1)  # horizontal
```

# 10. APPLY & MAP:

```python
# apply function to column
df["marks"] = df["marks"].apply(lambda x: x*1.1)

# apply to entire dataframe
df.apply(lambda x: x.max())

# map values
df["grade"] = df["marks"].map({
    85:"A", 92:"A", 78:"B", 88:"A", 65:"C"
})

# string operations
df["name"].str.upper()
df["name"].str.lower()
df["name"].str.len()
df["name"].str.contains("R")
```

# 11. PIVOT TABLE:

```python
df = pd.DataFrame({
    "name"  : ["Rahul","Priya","Rahul","Priya"],
    "subject":["Math","Math","Science","Science"],
    "marks" : [85, 92, 78, 88]
})

pivot = pd.pivot_table(
    df,
    values ="marks",
    index  ="name",
    columns="subject",
    aggfunc="mean"
)
print(pivot)

```
