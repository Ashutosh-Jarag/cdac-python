```python
arr = np.random.randint(1,17,16)
print(arr)

def isPrime(n):
    if n < 2:
        return False          
    for i in range(2, int(n**0.5)+1):  
        if n % i == 0:
            return False      
    return True               

p = [x for x in arr if isPrime(x)]
print(p)
```
