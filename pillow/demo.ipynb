```python
# img = Image.open("https://picsum.photos/id/237/200/300")
# img.show()

import requests
from PIL import Image
from io import BytesIO

url = "https://picsum.photos/id/237/200/300"
response = requests.get(url)
img = Image.open(BytesIO(response.content))

img.show()
```

```python
display(img)
```

```python
print(img.size)
print(img.height)
print(img.width)
print(img.filename)
print(img.mode)
print(img.format)
```

```python
img2=img.resize((300,300))
display(img2)
```

```python
img2.rotate(90)
```
